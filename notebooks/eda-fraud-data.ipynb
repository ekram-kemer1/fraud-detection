```python
ip_df = pd.read_csv("../data/raw/IpAddress_to_Country.csv")
```

```python
df["ip_address"] = df["ip_address"].astype(float)
```

```python
ip_df["lower_bound_ip_address"] = ip_df["lower_bound_ip_address"].astype(float)

ip_df["upper_bound_ip_address"] = ip_df["upper_bound_ip_address"].astype(float)
```

```python
df = df.sort_values("ip_address")
ip_df = ip_df.sort_values("lower_bound_ip_address")
```

```python
merged_df = pd.merge_asof(
    df,
    ip_df,
    left_on="ip_address",
    right_on="lower_bound_ip_address",
    direction="backward"
)
```

```python
merged_df = merged_df[
    merged_df["ip_address"]
    <= merged_df["upper_bound_ip_address"]
]
```

```python
merged_df["country"].value_counts().head(10)
```

```python
country_fraud = pd.crosstab(
    merged_df["country"],
    merged_df["class"]
)
```

```python
country_fraud.head(10).plot(kind="bar")
plt.show()
```
